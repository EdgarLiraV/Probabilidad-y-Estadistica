```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
import math
```

```python
Autos = pd.read_csv("autos2.csv", encoding = "ISO-8859-1")
```

```python
Autos.info()
```

##Cargar el Dataset de Autos y realizar las Estadísticas descriptivas básicas y medidas de tendencia central


```python
Valor_minimo = Autos.price.min()
Valor_Maximo = Autos.price.max()
Moda = Autos.price.mode()
Media = Autos.price.mean()
Mediana = Autos.price.median()
Desviacion_Estandar = Autos.price.std()
Estadisticas = Autos.price.describe()
print("Valor minimo: ", Valor_minimo, "\n Valor maximo: ", Valor_Maximo, "\n Moda: ", Moda, "\n Media: ", Media, "\n Mediana: ",Mediana, "\n Desviacion_Estandar: ", Desviacion_Estandar, "\n Estadisticas: ", Estadisticas)
```

##Realizar un Histograma de precios de los automóviles

Conclusiones:

Mientras mas caros son los carros menos frecuencia tienen, aunque hay un precio alto que se repite mas que incluso algunos baratos que es 35000

```python
k = int (1+ math.log2(len(Autos.price)))
plt.hist(Autos['price'], bins=k, alpha = 0.5)
plt.title("Histograma de precios de los automoviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

##Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo de vehículo (body style)
Conclusiones:

Me parece curioso la diferencia de precios entre todos, habiendo claramente 3 modelos economicos y 2 muy caros, aunque uno de los economicos tiene variables caras




```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='body-style', y='price', data=Autos)
plt.title('Relación entre Precio y Tipo de Vehículo')
plt.xlabel('Tipo de Vehículo (Body Style)')
plt.ylabel('Precio')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y tipo tracción ruedas (drive wheels)

Conclusiones:

Hay muy opcos coches con 4wd, los coches mas bratos en promedio son los fwd y los mas caros los rwd

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='drive-wheels', y='price', data=Autos)
plt.title('Relación entre Precio y traccion de ruedas')
plt.xlabel('Tipo de traccion de ruedas')
plt.ylabel('Precio')
plt.show()
```

##Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automóvil (price) y la ubicación del motor (engine-location)
Conclusiones:

El diagrama es un poco raro, ya que pocas veces el motor se encuentra en la parte trasera y su precio es muy alto, mientras que en la delantera hay muchos datos valores que sobresalen aunque su promedio sea bajo de precio

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='engine-location', y='price', data=Autos)
plt.title('Relación entre Precio y Ubicación del Motor')
plt.xlabel('Ubicación del Motor (Engine Location)')
plt.ylabel('Precio')
plt.show()
```
